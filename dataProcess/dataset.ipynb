```python
data_path = '/home/ubuntu/data/dataset/ml-100k/'
UIT2 = pd.read_csv(data_path + 'ml.trace',header=None)
UIT2
```

```python
trainUIT = UIT[UIT['day']<max(UIT['day']+1)*0.6]
validUIT = UIT[UIT['day']<max(UIT['day']+1)*0.8]
contentNum = len(UIT.i.drop_duplicates())
userNum = len(UIT.u.drop_duplicates())
contentNum,userNum,trainUIT,validUIT
```

```python
uWatch = []
for userId in range(userNum):
    uWatch.append(UIT[UIT.u==userId].i.unique())
    if len(uWatch[userId])<20:
        print(userId,uWatch[userId])
```

```python
def jaccard(x,y):
    intersection = set(x)&set(y)
    union = set(x)|set(y)
    return len(intersection)/len(union)
```

```python
userJaccard = np.zeros((userNum,userNum))
for userId in range(userNum):
    for userId2 in range(userId+1,userNum):
        userJaccard[userId,userId2] = jaccard(uWatch[userId], uWatch[userId2])
        userJaccard[userId2,userId] = userJaccard[userId,userId2]
```

```python
simUser = np.zeros((userNum,int(userNum*0.03)),dtype=int)
for userId in range(userNum):
    simUser[userId] = userJaccard[userId,:].argsort()[::-1][0:int(userNum*0.03)]
simUser[0]
```

```python
data_path = '/home/ubuntu/data/dataset/ml-100k/'

np.savetxt(data_path+"simUser.csv", simUser, fmt='%d',delimiter=',')
```
